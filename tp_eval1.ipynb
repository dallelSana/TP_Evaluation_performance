## Objectif du TP:

<ol>
<li>Estimer les paramtres empirique (la moyenne et la variance)</li>
<li>Faire varier le nombre d'chantillons</li>
<li>Vrifier la concordance entre moyenne thorique/empirique et variance thorique/empirique</li>
</ol>

Considrons une variable alatoire X suivant une loi f dans une population donne.
n-chantillon: chantillon de taille n regroupant un ensemble de variables alatoires
indpendantes et identiquement distribues, de mme loi f que X.
En pratique: (X$_{1}$,X$_{2}$,....X$_{n}$) == observations ralises 

avec $\bar{X}$=  $\frac{1}{n}$ $\sum_{i=1}^{n} X_i$ : Moyenne empirique <br>
$\sigma^{2}$= $\frac{1}{n}$ $\sum_{i=1}^{n} (X_i$ -$\bar{X}$)<sup>2</sup> : Variance empirique 


# exemple 1 :

si X a la distribution uniforme continue sur l'intervalle [0,1] , on a :<br>
$\mathrm{EX}=\frac{1}{2}, \quad \sigma_{X}^{2}=\frac{1}{12}, \quad F_{X}(x)=x, \quad x \in[0,1]$

```python

"""
    Importation des bibliothques ncessaires
        
"""

import random # Permet la gnration de nombres alatoires
import matplotlib.pyplot as plt #Chaque fonction pyplot modifie une figure:
                                #Exmple: cre une figure, cre une zone de traage dans une figure, etc.


"""
    Calcul de la moyenne empirique
        
"""
def moyenne(t,n): #Dclaration de la fonction moyenne 
    x=0
    for j in t: #calcul de la moyenne empirique
        x=x+j
    return x/n
"""
    Calcul de la variance empirique
        
"""
def variance(t,n,moyenne): #Dclaration de la fonction variance
    ta=0
    for i in t:
        ta=ta+(i-moyenne)**2 ## Calcul de la variance empirique en utilisant la fonction de la moyenne
    return ta/n

"""
    cration des valeurs alratoires
        
"""

def table(n):
    t=[] #liste vide
    for i in range(n+1): #Cration d'une squence de nombres de 0  n
        t.append(random.uniform(0, 1)) #Prendre des valeurs alatoires de i compris entre 0 et 1 
                                       #et les ajoutant dans la liste t 
    return t
n=[10,100,1000,10000] #les chantillons qu'on va utiliser
teta=[]
xn=[]
for i in n:
    t=table(i) #appel de la fonction tabe qui contient les valeurs alatoires
    xn.append(moyenne(t,i)) #xn va prendres les moyennes empiriques pour chaque chentillon
    teta.append(variance(t,i,moyenne(t,i))) #xn va prendres les variances empiriques pour chaque chentillon

for i in range(len(n)):
    print ('Echantillon= {} ,Moyenne_empirique= {} ,Variance_empirique= {}.'  
           .format(n[i],xn[i],teta[i])) #affichage des valeurs appels
print("Moyenne thorique = ",1/2)
print("Variance thorique = ",1/12)
```

```python
"""
    Reprsentation graphique des fonctions
        
"""

plt.plot(n,teta,'g+',label="Moyenne emprique") #trace des moyennes dans une zone de traage
plt.plot(n,xn,'b+', label="Variance emprique") #trace des variances dans une zone de traage
plt.xlabel("Nombre d'echantillon") #Affichage de l'tiquette de l'axe x
plt.ylabel("Variables alatoires") #Affichage de l'tiquette de l'axe y
plt.title("Moyenne emprique & variance emprique") #Ajout d'un titre au graphe
plt.legend()
plt.grid() #Affichage de la grille
plt.show() #Affichage de la courbe
```

## Exemple 2:

si X a la distribution exponentielle de paramtre $\lambda$ >0 , on a :
$\DeclareMathOperator{\e}{e}$<br>
$\mathrm{E} X=\frac{1}{\lambda}, \quad \sigma_{X}^{2}=\frac{1}{\lambda^{2}}, \quad F_{X}(x)=1-e^{-\lambda x}, \quad x \geq 0$

```python
"""
    Importation des bibliothques ncessaires
        
"""

import numpy as np #NumPy est une extension du langage de programmation Python, destine  manipuler
                    #des matrices ou tableaux multidimensionnels
                    #ainsi que des fonctions mathmatiques oprant sur ces tableaux. 
import math #Bibliothque pour les fonctions mathmatiques

"""
    Dclaration de la fonction
        
"""
def table(n,lamda): #fonction qui retourne des valeurs alatoires
    t=np.random.exponential(1/lamda,n)
    return t

"""
    Reprsentation graphique des fonctions
        
"""
xn=[]
teta=[]
lamda=1000 # les chantillons de lemda qu'on va utiliser

for i in range(4):
    t=table(n[i],lamda) #t va prendre les valeurs alatoires
    x=moyenne(t,n[i])
    xn.append([x]) #Ajout des moyennes empiriques dans Xn
    teta.append([variance(t,n[i],x)]) #Ajout des variances empiriques dans teta
    plt.subplot(2,2,1) #subplot permet d'organiser diffrents tracs  l'intrieur d'une grille d'affichage
    plt.plot(n[i],teta[i],'g+')
    plt.xlabel("lemda")
    plt.ylabel("variance")
    plt.subplot(2,2,2)
    plt.plot(n[i],x,'b+')
    plt.xlabel("lemda")
    plt.ylabel("moyenne")
plt.show()
#for j in range(len(n)):
for i in range(len(n)):
    print('Le variable aleatoire Xi est= {} ,la moyenne_empirique= {} ,la variance_empirique= {}'.format(n[i],xn[i],teta[i]))
print("Moyenne thorique = ",1/lamda)
print("Variance thorique = ",1/lamda**2)
```

## Exemple 3:

si X a la distribution discrte uniforme {a,a+1,.....,b-1,b} , a , b $\in$ $\mathbb N$  et a $\leq$ b ,alors :

$\mathrm{E} X=\frac{a+b}{2} \quad \sigma_{X}^{2}=\frac{(a-b)(a-b-2)}{12}$ , $F_{X}(x)=\left\{\begin{array}{ll}0 & x<a \\ \frac{1+\mid x-a\rfloor}{b-a+1} & a \leq x<b \\ 1 & x \geq b\end{array}\right.$

```python
def table(n,a,b):
    return np.random.randint(a,b,n) #renvoie des valeurs alatoires
xn=[]
teta=[]
a=0
b=10
for i in n:
    t=table(i,a,b)
    x=moyenne(t,i)
    xn.append(x) #Attribuer le moyenne empirique vers xn
    teta.append(variance(t,i,x)) #Attribuer le variance empirique vers teta
plt.subplot(2,2,1)
plt.plot(n,teta,'g+',label="Moyenne")
plt.grid()
plt.subplot(2,2,2)
plt.plot(n,xn,'b+',label="Variance")
plt.grid()
plt.show()
for i in range(len(n)):
    print('Echantillon= {} ,Moyenne_empirique= {} ,Variance_empirique= {}.'.format(n[i],xn[i],teta[i]))
print("Moyenne thorique = ",(a+b)/2)
print("Variance thorique = ",((a-b)*(a-b-2))/12)
```

## Exemple 4:

si X  a la distribution discrete telle que $$
\mathbb{P}\{X=0\}=0,3 \quad \mathbb{P}\{X=2\}=0,2 \quad \mathbb{P}\{X=6\}=0,5
$$

alors:
$$
\mathbb{E} X=3,4 \quad \sigma_{X}^{2}=7,24
$$
et
$$
F_{X}(x)=\left\{\begin{array}{ll}
0 & x<0 \\
0,3 & 0 \leq x<2 \\
0,5 & 2 \leq x<6 \\
1 & x \geq 6
\end{array}\right.
$$




```python
def table(n):
    x=[]
    t=np.random.uniform(0, 1,n)
    for i in t:
        if i<0.3:
            x.append(0)
        elif i<0.5:
            x.append(2)
        else:
            x.append(6)
    return x
xn=[]
teta=[]
for i in n:
    t=table(i)
    x=moyenne(t,i)
    xn.append(x)
    teta.append(variance(t,i,x))
for i in range(len(n)):
    print ('Echantillon= {} ,Moyenne_empirique= {} ,Variance_empirique= {}.'.format(n[i],xn[i],teta[i]))
    
plt.subplot(2,2,1)
plt.plot(n,teta,'g+',label="Moyenne")
plt.grid()
plt.subplot(2,2,2)
plt.plot(n,xn,'b+',label="Variance")
plt.grid()
plt.show()

print("Moyenne thorique = ",3.4)
print("Variance thorique = ",7.24)
```
